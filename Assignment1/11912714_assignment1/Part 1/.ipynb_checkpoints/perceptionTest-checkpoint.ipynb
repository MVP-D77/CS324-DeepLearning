```python
import numpy
from perceptron import *
import matplotlib.pyplot as plt
```

```python
x1,x2,y1,y2 = generate(2,2,7,2)
```

```python
plt.scatter(x1[0:200,0],x1[0:200,1], color='b', marker='*', label='positive')
plt.scatter(x2[0:200,0],x2[0:200,1], color='r', marker='*', label='negative')
plt.legend(loc='upper left')
plt.title('Guassian distribution DataSet')
plt.show()
```

```python
x_train = np.concatenate((x1[:160],x2[:160]), axis = 0)
y_train = np.concatenate((y1[:160],y2[:160]), axis = 0)
x_test = np.concatenate((x1[160:],x2[160:]), axis = 0)
y_test = np.concatenate((y1[160:],y2[160:]), axis = 0)
x_train.shape
```

```python
perceptron = Perceptron(320)
perceptron.train(x_train, y_train, x_test, y_test)
```

```python
test_accu = perceptron.get_test_accu()
print('The final classification accuracy on the test set is',test_accu[-1])
```

```python
x_axis = [i for i in range(1, len(test_accu) + 1)]
plt.ylim(0,1)
plt.plot(x_axis, test_accu, c='red', label='training accuracy')
plt.legend(loc='upper left')
plt.xlabel('epoch')
plt.ylabel('test_accurate_rate')
plt.show()
```

```python
m1 = ini_m2 = 2
perceptron = Perceptron(320)
accu_change = []
for delta in range(10):
    m2 = ini_m2+delta
    x1,x2,y1,y2 = generate(m1,2,m2,2)
    x_train = np.concatenate((x1[:160],x2[:160]), axis = 0)
    y_train = np.concatenate((y1[:160],y2[:160]), axis = 0)
    x_test = np.concatenate((x1[160:],x2[160:]), axis = 0)
    y_test = np.concatenate((y1[160:],y2[160:]), axis = 0)
    
    perceptron.train(x_train, y_train, x_test, y_test)
    accu_change.append(perceptron.get_test_accu()[-1])
```

```python
x_axis = [i for i in range(1, len(accu_change) + 1)]
plt.ylim(0,1.1)
plt.plot(x_axis, accu_change, c='red', label='training accuracy')
plt.legend(loc='upper left')
plt.xlabel('epoch')
plt.ylabel('test_accurate_rate')
plt.show()
```

```python
s1 = ini_s2 = 2
perceptron = Perceptron(320)
accu_change = []
for delta in range(10):
    s2 = ini_s2+delta
    x1,x2,y1,y2 = generate(2,s1,10,s2)
    x_train = np.concatenate((x1[:160],x2[:160]), axis = 0)
    y_train = np.concatenate((y1[:160],y2[:160]), axis = 0)
    x_test = np.concatenate((x1[160:],x2[160:]), axis = 0)
    y_test = np.concatenate((y1[160:],y2[160:]), axis = 0)
    
    perceptron.train(x_train, y_train, x_test, y_test)
    accu_change.append(perceptron.get_test_accu()[-1])
```

```python
x_axis = [i for i in range(1, len(accu_change) + 1)]
plt.ylim(0,1.1)
plt.plot(x_axis, accu_change, c='red', label='training accuracy')
plt.legend(loc='upper left')
plt.xlabel('epoch')
plt.ylabel('test_accurate_rate')
plt.show()
```
